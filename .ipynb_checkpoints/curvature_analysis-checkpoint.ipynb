```python
#import modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#import data
data = pd.read_csv("raw_combined.csv")
print(f"MoCap Data Size: {data.shape}")
data.head()
```

### Data Cleaning

```python
#Delete rows we know we don't need
data.drop(["shotNo"],axis=1,inplace=True)
```

```python
#replace club names with labels
club_mapping = {"X026": 0, "TS2": 1}
data['headID'] = data['headID'].map(club_mapping)
```

```python
print(data.shape)
data.head()
```

```python
#dealing with NaN values

#1 find how many NaN values in each row
NaN_series = data.isnull().sum(axis=1).sort_values(ascending=False)

#find a good threshold of rows to just delete completely
#final: delete all rows with more than 10 NaN values (coverage too poor)
z = 0
thresh = 5
bad_rows = NaN_series[(NaN_series > z) == True].shape
worst_rows = NaN_series[(NaN_series > thresh) == True].shape
print(f"Number of Rows w NaN Values: {bad_rows[0]}")
print(f"Number of Rows w more than {thresh} NaN Values: {worst_rows[0]}")
```

```python
#delete rows that have NaN values above this threshold
data_drop = NaN_series[(NaN_series > thresh) == True].index
data = data.drop(data.index[data_drop])
print(data.shape)
data.head()
```

```python
#fill NaN values with approximated values
column_names = list(data)
for i in range(1,len(data.columns)):
    if (data.iloc[:,i].isna().any() == True):
        x = data.groupby("playerID")[column_names[i]].transform('median')
        #fill empty rows with the corresponding median value in this pandas series
        data[column_names[i]].fillna(x, inplace=True)
```

### Exploratory Data Analysis

```python
list(data)
```

```python
#Big Picture before data parsing

#total number of players
players_num=len(np.unique(data['playerID']))
print(f"number of players: {players_num}")

#total number of shots
shots_num=data.shape[0]
print(f"total shots: {shots_num}")

avg_data = pd.DataFrame(round(data.groupby('headID')['Launch Direction', 'Side', 'Curvature', 'Face'
                                                     , 'EffFace','ExFace','ImpactX', 'Path'].mean(),2))
avg_data
```

### Multilinear Regression Model on All Data

```python
from sklearn.model_selection import train_test_split

train_data, test_data = train_test_split(data, test_size=.2)
print(train_data.shape) 
print(test_data.shape)
```

```python
#MULTILINEAR REGRESSION (all variables)
from sklearn import linear_model
import statsmodels.api as sm
from sklearn import metrics

X = train_data.iloc[:,13:]
y = train_data['Side']

ml_reg = linear_model.LinearRegression()
ml_reg.fit(X,y)

model = sm.OLS(y, X).fit()
predictions = model.predict(X)
model.summary()
```

```python
#MULTILINEAR REGRESSION (certain variables)
from sklearn import linear_model
import statsmodels.api as sm
from sklearn import metrics

X = train_data[['Face', 'Path', 'ExFace', 'ImpactX']]
y = train_data['Side']

ml_reg = linear_model.LinearRegression()
ml_reg.fit(X,y)

model = sm.OLS(y, X).fit()
predictions = model.predict(X)
model.summary()
```

```python
#Error between predictions and actual (absolute error)
X_test = test_data[['Face', 'Path', 'ExFace', 'ImpactX']]
y_pred = ml_reg.predict(X_test)
print(np.sqrt(metrics.mean_absolute_error(test_data["Side"], y_pred)))
```
